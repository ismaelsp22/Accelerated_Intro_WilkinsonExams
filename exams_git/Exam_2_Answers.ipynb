```python
import re

def locus_code(file, lista):
    open_file =  open(file, "r")
    n = 0
    for line in open_file:
        if n == 0:
            n += 1
        else:
            mo= re.search( r'AT\dG\d{5}', line)
            lista.append(mo.group())
        
def comparation_locus(list1, list2):
    n = 0
    for position in list1:
        if position == list2[n]:
            print("Same sequence")
        else:
            print("Not same sequence")
        n += 1

A = []
locus_code("Germplasm.tsv", A)      
B = []
locus_code("LocusGene.tsv", B)

comparation_locus(A,B)
```

```python
import pymysql.cursors
import csv
connection = pymysql.connect(host='localhost',
                            user='root',
                            password='root',
                            db='exam_database',
                            charset='utf8mb4', 
                            cursorclass=pymysql.cursors.DictCursor)
connection.autocommit = False

try:
    with connection.cursor() as cursor:
        with open('Germplasm.tsv') as csvfile:
            germp = csv.DictReader(csvfile, delimiter = "\t")
            for row in germp:
                print(row)
                sql = """INSERT INTO germplasm (Locus, germplasm, phenotype, pubmed) values ('{}', '{}', '{}', '{}')""".format(row["Locus"], row["germplasm"], row["phenotype"], row["pubmed"])
                cursor.execute(sql)
        connection.commit() 


finally:
    connection.close() 
    
```
